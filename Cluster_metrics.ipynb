```python
data_dir = "ngrams/timeseries_data/"
file_name = data_dir + "books_per_year"
bookcount = pd.read_csv(file_name, delim_whitespace = True, header = None)
bookcount.columns = ['_year', 'total_books']
bookcount.sort(columns = '_year', inplace = True)
bookcount['_year'] = bookcount['_year'].apply(int)
bookcount['total_books'] = bookcount['total_books'].apply(int)
bookcount = bookcount.set_index(bookcount['_year'])
bookcount = bookcount[bookcount['_year'] >= 1800]
bookcount = bookcount[bookcount['_year'] <= 2012]
```

```python
normalized_by_total_books = allwords.astype(float).div(bookcount['total_books'], axis = 0)
normalized_by_total_books.fillna(0)
```

```python
dmin, dmax = normalized_by_total_books.min(axis=0), normalized_by_total_books.max(axis=0)
normalized_by_word = (normalized_by_total_books - dmin) / (dmax - dmin)
```

```python
def skewness(series):
    m = series.mean()
    s = series.std()
    diff = series.sub(m)
    cubed = diff.apply(lambda x: x**3)
    return (cubed.sum())/(len(series)*s**3)
```

```python
def kurtosis(series):
    m = series.mean()
    s = series.std()
    diff = series.sub(m)
    cubed = diff.apply(lambda x: x**4)
    return ((cubed.sum())/(len(series)*s**4))-3
```

Agglomerative Clustering
========================

```python
from scipy.spatial.distance import pdist, squareform
#This matrix is condensed, to get the redundant matrix call squareform(distMatrix)
distMatrix = pdist(metrics_norm[['Kurtosis', 'Variance', 'Skewness']])
```

```python
squareform(distMatrix)[0][1]
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
linked = linkage(distMatrix, method='complete')
#linked = linkage(metrics_norm[['Kurtosis', 'Variance', 'Skewness']], method='centroid')
dendrogram(linked,         color_threshold=0.3,            leaf_font_size=6)
```

```python
hclusters = DataFrame(fcluster(linked, 1.15470053), metrics_norm.index, columns = ['Hierarchical_Cluster'])
```

```python
hier_df = pd.concat([metrics_norm, hclusters], axis = 1)
```
