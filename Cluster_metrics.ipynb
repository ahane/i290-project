```python
data_dir = "ngrams/timeseries_data/"
file_name = data_dir + "books_per_year"
bookcount = pd.read_csv(file_name, delim_whitespace = True, header = None)
bookcount.columns = ['_year', 'total_books']
bookcount.sort(columns = '_year', inplace = True)
bookcount['_year'] = bookcount['_year'].apply(int)
bookcount['total_books'] = bookcount['total_books'].apply(int)
bookcount = bookcount.set_index(bookcount['_year'])
bookcount = bookcount[bookcount['_year'] >= 1800]
bookcount = bookcount[bookcount['_year'] <= 2012]
```

```python
bookcount.head()
```

```python
normalized_by_total_books = allwords.astype(float).div(bookcount['total_books'], axis = 0)
normalized_by_total_books.fillna(0)
```

Normalize every word timeseries to be in range [0;1]

```python
dmin, dmax = normalized_by_total_books.min(axis=0), normalized_by_total_books.max(axis=0)
normalized_by_word = (normalized_by_total_books - dmin) / (dmax - dmin)
```

```python
normalized_by_word = normalized_by_word.fillna(0)
```

```python
def plot_regress_line(word):
    from scipy import stats
    #slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    united = stats.linregress(normalized_by_word[word].index, normalized_by_word[word])
    regress = map(lambda x: united[0]*x + united[1], range(1800, 2013))
    regser = Series(regress, index = range(1800, 2013))
    #normalized_by_word[word].plot()
    #normalized_by_word[word].plot()
    substracted = (normalized_by_word[word]-regser)
    return substracted
```

```python
def skewness(series):
    m = series.mean()
    s = series.std()
    diff = series.sub(m)
    cubed = diff.apply(lambda x: x**3)
    return (cubed.sum())/(len(series)*s**3)
```

```python
def kurtosis(series):
    m = series.mean()
    s = series.std()
    diff = series.sub(m)
    cubed = diff.apply(lambda x: x**4)
    return ((cubed.sum())/(len(series)*s**4))-3
```

```python
from scipy import stats
agg_df = {'Word': [], 'Mean': [], 'Variance': [], 'Skewness': [], 'Kurtosis': [], 'Reg_Slope': [], 'Reg_Error': [], 'r': []}

for word in normalized_by_word.columns:
    agg_df['Word'].append(word)
    agg_df['Mean'].append(normalized_by_word[word].mean())
    agg_df['Variance'].append(normalized_by_word[word].var())
    agg_df['Skewness'].append(skewness(normalized_by_word[word]))
    agg_df['Kurtosis'].append(kurtosis(normalized_by_word[word]))
    slope, intercept, r_value, p_value, std_err = stats.linregress(normalized_by_word[word].index, normalized_by_word[word])
    agg_df['Reg_Slope'].append(slope)
    agg_df['Reg_Error'].append(std_err)
    agg_df['r'].append(r_value)
```

```python
metrics = DataFrame(agg_df)
```
