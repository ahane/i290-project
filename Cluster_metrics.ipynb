```python
def skewness(series):
    m = series.mean()
    s = series.std()
    diff = series.sub(m)
    cubed = diff.apply(lambda x: x**3)
    return (cubed.sum())/(len(series)*s**3)
```

```python
def kurtosis(series):
    m = series.mean()
    s = series.std()
    diff = series.sub(m)
    cubed = diff.apply(lambda x: x**4)
    return ((cubed.sum())/(len(series)*s**4))-3
```

```python
from scipy import stats
agg_df = {'Word': [], 'Mean': [], 'Variance': [], 'Skewness': [], 'Kurtosis': [], 'Reg_Slope': [], 'Reg_Error': [], 'r': [], 'p_value': []}

for word in normalized_by_word.columns:
    agg_df['Word'].append(word)
    agg_df['Mean'].append(normalized_by_word[word].mean())
    agg_df['Variance'].append(normalized_by_word[word].var())
    agg_df['Skewness'].append(skewness(normalized_by_word[word]))
    agg_df['Kurtosis'].append(kurtosis(normalized_by_word[word]))
    slope, intercept, r_value, p_value, std_err = stats.linregress(normalized_by_word[word].index, normalized_by_word[word])
    agg_df['Reg_Slope'].append(slope)
    agg_df['Reg_Error'].append(std_err)
    agg_df['r'].append(r_value)
    agg_df['p_value'].append(p_value)
```

```python
import scipy.cluster.vq as vq
def cluster_timeseries(features_df, timeseries_df, features = None, k = 5):
    if features:
        whitened = vq.whiten(features_df[features])
    else:
        whitened = vq.whiten(features_df)
    book = array(whitened)
    centroids, kmeans_labels = vq.kmeans2(book, k=5)
    kmclusters = DataFrame(kmeans_labels, features_df.index, columns = ['KMeans_Cluster'])
    kmeans_df = pd.concat([features_df, kmclusters], axis = 1)
    
    return kmeans_df
```

```python
def plot_cluster_means(kmeans_df, cluster_column = 'KMeans_Cluster'):
    fig = plt.figure(figsize = (15, 10))
    #c = kmeans_df[cluster_column].value_counts()
    for label, df in kmeans_df.groupby(cluster_column):
        sp = plt.subplot(3, 4, label+1)
        n = len(df)
        sp.text(0.95, 0.95,'c= %s n=%s'%(str(label), str(n)),
                horizontalalignment='right',
                verticalalignment='top',
                transform=sp.transAxes, 
                fontsize = 14)
        words = list(df.index)
        timeseries = normalized_by_total_books[words]
        timeseries.mean(axis = 1).plot()
```
