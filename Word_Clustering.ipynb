Loading and Preprocessing the Word Occurance Timeseries
-------------------------------------------------------

```python
import re
import json
import codecs
```

```python
words_ts_tup = []
with codecs.open(WORD_TIMESERIES, encoding='utf-8') as f:
    for line in f:

        m = re.match(r'\"([\w\\]+)\"\s([\[\]\w\s\,]+)', line)
        if m:

            key = m.group(1)
            value = json.loads(m.group(2))
            total = value[1]
            time_tuples = map(tuple, value[3])
            if total > MIN_TOTAL_OCC:
                words_ts_tup.append( (key, time_tuples, total) )
        else:

            print 'regex error'
```

```python
words_ts_df = DataFrame(words_ts_tup, columns = ['word', 'ts', 'total'])
del words_ts_tup
```

```python
words_ts_df.head()
```

We now have to convert the list of tuples in the 'ts' column into pandas objects

```python
#This function creates a list of single-column dataframes to concatenate later
timeseries_dfs = []
def mk_clean_series(row):
    if row['total'] >= MIN_TOTAL_OCC:
        word = row['word']
        ts = row['ts']
        word_df = DataFrame.from_records(ts, columns = ['_year', word])
        word_df = word_df[word_df['_year'] <= MAX_YEAR]
        word_df = word_df[word_df['_year'] >= MIN_YEAR]
        word_df['_year'] = word_df['_year'].map(int)
        word_df = word_df.set_index('_year')
        timeseries_dfs.append(word_df)
    return None
```

```python
words_ts_df.apply(mk_clean_series, axis = 1)
```

```python
words_ts_abs = pd.concat(timeseries_dfs, axis = 1)
words_ts_abs = words_ts_abs.fillna(0)
```

Choice of Features
------------------
To quantify 'interesting' we came up with the following features:

- Length of Longest Streak
- Last Year of Longest Streak

    A streak is defined as a time period in which the timeseries is above it's mean. A long streak therefore indicates that a word has changed its occurance rate substantially for a longer period, something we can see in 'development' but not in 'make'. The absence of long streaks indicates high fluctuation rates.
    
    The last year of a the longest streak is of interest, as this allows us to seperate words for which the longest streak is a peak, from words that are still (ie. in 2011) in their longest streak.
    

```python
def longest_over_mean(series):
    """Returns the last year and the length of the longest period in which a timeseries was above mean"""
    mean = series.mean()
    min_year = min(series.index)
    max_year = max(series.index)
    streaks = []
    in_streak = False
    for year, value in zip(series.index, series):
        if value >= mean:
            if in_streak:
                streaks[-1].append(year)
            else:
                in_streak = True
                streaks.append([year])
        if value < mean:
            in_streak = False
    length_last = map(lambda x: (len(x), x[-1]), streaks)
    return max(length_last)
```

- Median Year
    
    This is the year in which half of the sum of all relative occurances has been reached. This can be interpreted as the center of weight of a timeseries, and thus is a metric to group words with peaks in time.

```python
def median_year(series):
    half = series.sum()/2
    count = 0
    min_year = min(series.index)
    max_year = max(series.index)
    for year, value in zip(series.index, series):
        count += value
        if count >= half:
            return year
```

Apart from these self-written metrics we use:

- Skewness

    To find words which only occure only in limited time periods, ie. have lots of counts near zero and a few very high counts

- Regression Slope

    An naive way to seperate upwards trending from downwards trending words


Creating the Features DataFrame
-------------------------------

```python
from scipy import stats
features_dict = {'Word': [], 'Skewness': [],  'Reg_Slope': [], 
                 'longest_streak': [], 'last_year_of_streak': [], 'median_year': []}

for word in words_ts_rel_mm.columns:
    features_dict['Word'].append(word)

    features_dict['Skewness'].append(words_ts_rel_mm[word].skew())
    
    
    slope, intercept, r_value, p_value, std_err = stats.linregress(words_ts_rel_mm[word].index, words_ts_rel_mm[word])
    features_dict['Reg_Slope'].append(slope)
    
    length, last_year_of_streak = longest_over_mean(words_ts_rel_mm[word])
    features_dict['longest_streak'].append(length)
    features_dict['last_year_of_streak'].append(last_year_of_streak)
    
    features_dict['median_year'].append(median_year(words_ts_rel_mm[word]))
    
```

K-Means Clustering
------------------
We define functions for clustering and convenient plotting of the results

```python
import scipy.cluster.vq as vq
def cluster_timeseries(features_df, features = None, num_clusters = 5):
    if features:
        whitened = vq.whiten(features_df[features])
    else:
        whitened = vq.whiten(features_df)
    book = array(whitened)
    centroids, kmeans_labels = vq.kmeans2(book, k=num_clusters)
    kmclusters = DataFrame(kmeans_labels, features_df.index, columns = ['KMeans_Cluster'])
    kmeans_df = pd.concat([features_df, kmclusters], axis = 1)
    return kmeans_df
```

```python
def plot_cluster_mean_ts(kmeans_df, timeseries_df, cluster_column = 'KMeans_Cluster'):
    fig = plt.figure(figsize = (15, 10))
    for label, df in kmeans_df.groupby(cluster_column):
        sp = plt.subplot(3, 4, label+1)
        n = len(df)
        sp.text(0.95, 0.95,'c= %s n=%s'%(str(label), str(n)),
                horizontalalignment='right',
                verticalalignment='top',
                transform=sp.transAxes, 
                fontsize = 14)
        words = list(df.index)
        cluster_timeseries = timeseries_df[words]
        cluster_timeseries.mean(axis = 1).plot()
```
