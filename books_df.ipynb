```python
def mk_clean_series(year_counts):
    if year_counts:
        s = DataFrame.from_records(year_counts, columns =['year', 'count'])
        s = s.set_index('year')
        s = s[s['count'] <= MAX_YEAR]
        return s['count']
    else:
        return None
```
