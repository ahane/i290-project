```python
wc = open('wordcount.txt')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
w = pd.read_csv(wc, sep='\t')
```

```python
w.columns = ['Word', 'Count']
```

```python
s = w.sort(column = 'Count', ascending = False)
```

```python
s[:30]
```

```python
g = w.groupby('Count')
```

```python
wordcounts = []
occurance = []
for c, grp in g:
    #print c, grp['Word'].count()
    wordcounts.append(c)
    occurance.append(grp['Word'].count())
```

```python
wc_dist = pd.DataFrame({'Wordcount': wordcounts, 'Occurance': occurance})
```

```python
wc_dist
```

```python
plt.bar(wc_dist['Wordcount'], wc_dist['Occurance'])
plt.xlim(0, 350)
plt.show()
```
